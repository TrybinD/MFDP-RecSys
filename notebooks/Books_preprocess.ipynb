# Предобработка данных по книгам

```python
import sys
sys.path.append('../')

import pandas as pd
from pathlib import Path
from tqdm import tqdm
```

```python
books_ds_path = Path('../datasets/books/')
```

```python
df_books_rating = pd.read_csv(books_ds_path / 'Ratings.csv')
df_books = pd.read_csv(books_ds_path / 'Books.csv')
df_boooks_users = pd.read_csv(books_ds_path / 'Users.csv')
```

## 1. Сократим количество книг объединив книги с одинаковым названием и удалив непопулярные книги

```python
df_books[df_books['Book-Title'].str.lower().str.contains('fight club')]
```

```python
# Удалим безсмысленную приписку ': A Novel' в конце некоторых книг
df_books['Book-Title'] = df_books['Book-Title'].str.replace(r'(: [Aa] [Nn]ovel)', '', regex=True)
```

```python
df_books[df_books['Book-Title'].str.lower().str.contains('a novel')]
```

```python
df_books['Book-Title'] = df_books['Book-Title'].str.strip()
```

```python
df_books['Year-Of-Publication'] = df_books['Year-Of-Publication'].astype(int)
```

```python
df_books
```

```python
# Сгруппируем, если одинаковое название и автор. Остаим только среднее изображение, и первую публикацию,
# и все коды ISBN, чтобы настроить мэпинг объединенных книг

df_books = df_books.groupby(['Book-Title', 'Book-Author'], as_index=False).agg({'ISBN': lambda x: ', '.join(x), 
                                                    'Year-Of-Publication': min, 
                                                    'Image-URL-M': 'first'})
```

```python
df_books
```

```python
# Заменим все ISBN на первый + сохраним словарь для мэпинга рейтинга
isbn_list = []
map_dict = {}
for i in df_books['ISBN']:
    temp_isbn = i.split(', ')
    isbn_list.append('b'+temp_isbn[0])
    for j in temp_isbn:
        map_dict[j] = 'b'+temp_isbn[0]
```

```python
df_books['ISBN'] = isbn_list
```

```python
map_dict
```

```python
# Теперь датафрейм полностью готов к использованию, осталось только отчистить от непопулярных книг
df_books
```

```python
# Интересно, что в файле с рейтингами книг больше, чем в изначальном файле, странно. Удалим все, что не имеет пару.
len(map_dict), len(set(df_books_rating['ISBN']))
```

```python
df_books_rating['ISBN'] = df_books_rating['ISBN'].map(map_dict)
```

```python
df_books_rating = df_books_rating.dropna()
```

```python
# Довольно много книг, у которых меньше 2 рейтингов - почти половина. Их смело можно выкидывать
(df_books_rating['ISBN'].value_counts() < 2).sum()
```
